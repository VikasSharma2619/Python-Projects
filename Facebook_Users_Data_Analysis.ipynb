# **1) Load the data and impute missing values Imputation of missing values.**

           
---


 *● **Replace the null values (NA) of gender column with its mode or median and explain why mode/median used to replace NA values**.*







```python
data['gender'].isnull().sum()    ## checking the number of null values present
```

```python
data.gender.value_counts()       
```

```python
plt.figure(figsize=(7,7))
sns.set_style('darkgrid')
data['gender'].value_counts().plot.bar(color=['violet','indigo'], edgecolor='black',lw=2)     ##plotting bar chart
```

* We have 58,574 (i.e. 59.16%) males and only 40,254 (i.e. 40.66%) females and we don't have any other column (eg: Name) from which we can identify the gender of users.                    

* As we have also checked that in gender column, null (missing) values in raw data are only present in 175 rows (i.e. 0.17%) out of total 99003 rows, which will not have significant impact on the Model. 

* Hence, we can impute the missing values with mode (>50% or highest occuring value) of gender column i.e. "male" (59.16%)

```python
data['gender'] = data['gender'].fillna('male')        ##imputing missing values with mode value of gender column
```

---
*● **Replace the null values (NA) of tenure column (numerical variable) with its median, and explain why mode/median used to replace NA values**.*



```python
data['tenure'].isnull().sum()
```

```python
data['tenure'].mean(), data['tenure'].median()       ## checking the skewness with the help of median and mean value of tenure column
```

```python
plt.figure(figsize=(7,7))
sns.set_style('darkgrid')
data['tenure'].value_counts().sort_index().plot.area()    ## plotting 
```

* **To identify the distribution of the tenure column, I calculated mean (537.89) and median (412) which shows that mean > median. It means that the distribution is a positively skewed. And which is shown with the help of plotting above.** 
* **If we impute missing values of positively skewed data with mean, then outliers will increase the biasness of mean which will have the bad impact on model.**
* **As we have also checked that in tenure column, null (missing) values in raw data are only present in 2 rows (i.e. 0.002%) out of total 99003 rows, which will not have significant impact on the Model.**
* **That's why, I have imputed the missing values with median of gender column i.e. 412.**

```python
data['tenure'] = data['tenure'].fillna(data['tenure'].median())     ## imputing missing values with meidan of tenure column
```

*● **What is the distribution of tenure across different categories of gender**?*


**Ans: From the distribution plots, we understand the following things:**        
1. **The variaton of males are higher than females.**                       
2. **More males are using facebook for more than 500 and less than 1000 days as compared to females that's why density of male between this range is high (1st plot).**
3. **More females are using facebook for more than 1500 and less than 2250 days as compared males which can be observed in histplot (2nd plot).**

**Distribution(3rd) and line(4th) plot shows that although there are more males who use facebook for less than 500 days, but fall in facebook usage of males between 500-1000 days is higher as compared to those of females i.e. females use facebook more between 500-1000 days. That's why, a steep downward curve in distribution(1st) plot can be observed which is confirmed by other distribution(3rd) and line(4th) plot.**

```python
plt.figure(figsize=(7,7))              ## increasing the figure size for good visibility
sns.set_style('darkgrid')
sns.displot( data, x='tenure',col='gender', hue='gender', kind="kde", multiple="stack", palette='bright', height=8, edgecolor='black',lw=2)         ## plotting the distribution of tenure column
```

```python
plt.figure(figsize=(9,9))
sns.set_style('darkgrid')
sns.histplot(data=data, x='tenure', hue='gender', kde=True, cbar=True, palette= 'dark', lw=3)    ##histplot
```

```python
## subsetting the data to check the reason behind steep downward variation between 500-1000 in distribution plot
dumdata = data[(data['tenure']>500) & (data['tenure']<1000)]    

plt.figure(figsize=(7,7))        ## increasing th figure size for good visibility
sns.set_style('darkgrid')
sns.displot(dumdata, x='tenure', hue='gender', kind="kde", multiple="stack", palette='rainbow', height=8, edgecolor='black', lw=1)  ## plotting the distribution of tenure column
```

```python
plt.figure(figsize=(9,9))
sns.lineplot(y=dumdata.index, x=dumdata['tenure'], hue=dumdata['gender'], err_style='bars')    ## line plot between 500-1000 days
```
